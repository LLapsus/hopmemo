# Icon Dataset Demo

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from icon_dataset import generate_icons
```

```python
X, y = generate_icons(n=50, seed=42)   # (50,28,28)

print("X shape:", X.shape)
print("y shape:", y.shape)
print("Labels:", np.unique(y))
```

```python
fig, axs = plt.subplots(10, 5, figsize=(16, 30))    

for i in range(10):
    for j in range(5):
        index = i * 5 + j
        sns.heatmap(X[index], cmap='gray', cbar=False, ax=axs[i, j])
        axs[i, j].set_xticks([])
        axs[i, j].set_yticks([])
    
        title = f"Label: {y[index]}"
        axs[i, j].set_title(title, fontsize=16)

plt.show()
```
