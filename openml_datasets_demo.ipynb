# Dataset demo

Demo of loading dataset and plotting binarized samples.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from openml_datasets import *
```

## Dataset MNIST

```python
# Load MNIST dataset
ds_mnist = ImageDataset28().load(
    "mnist",
    bin_method="threshold",
    thr=30,
    limit=1000,
    seed=0,
)

# Get a stratified subset of 50 samples
X_mnist, y_mnist = ds_mnist.subset(50, stratify=True, seed=42)

print(f"Dataset name: {ds_mnist.name}")
print(f"Data shape: {X_mnist.shape}, Labels shape: {y_mnist.shape}")
print(f"Unique values in data: {np.unique(X_mnist)}")
```

```python
# Plot few sample
fig, axs = plt.subplots(1, 10, figsize=(30, 3))

for i in range(10):
    sns.heatmap(X_mnist[i], cmap="gray", cbar=False, ax=axs[i], xticklabels=False, yticklabels=False)
    axs[i].set_xticks([])
    axs[i].set_yticks([])
    
    title = f"Label: {y_mnist[i]}"
    axs[i].set_title(title, fontsize=16)
plt.show()
```
