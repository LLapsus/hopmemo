# Dataset demo

Demo of loading dataset and plotting binarized samples.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from dataset import *
```

```python
# Load Fashion-MNIST dataset
ds = ImageDataset28().load(
    "fashion-mnist",
    bin_method="threshold",
    thr=40,
    limit=1000,
    seed=0,
)

# Get a stratified subset of 50 samples
X, y = ds.subset(50, stratify=True, seed=42)

print(f"Data shape: {X.shape}, Labels shape: {y.shape}")
print(f"Unique values in data: {np.unique(X)}")
```
